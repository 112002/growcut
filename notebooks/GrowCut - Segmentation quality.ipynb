GrowCut - Segmentation experiments
----------------------------------

As described in:

    V. Vezhnevets, V. Konouchine. "Grow-Cut" - Interactive Multi-Label N-D Image Segmentation". 
    In Proceedings of the 2005 Conference, Graphicon. Pages 150 – 156.


```python
from growcut import growcut_cy
```

```python
import numpy as np
from matplotlib import pyplot as plt
from skimage import color, io, util
```

```python
import glob
import os
import collections

sample = collections.namedtuple('sample', ['image', 'mask'])

prefix = '/Users/nfaggian/Downloads/1obj'

samples = []
for sample_name in [os.path.split(x)[1] for x in glob.glob('{}/*'.format(prefix))]:
    if '.mat' in sample_name:
        continue
    img = io.imread('{0}/{1}/src_bw/{1}.png'.format(prefix, sample_name), as_grey=True)
    for mask_name in glob.glob('{}/{}/human_seg/*'.format(prefix, sample_name)):
        mask = io.imread(mask_name)[:,:,0] == 255
        break
    samples.append(sample(img, mask))
```

```python
index = 65

img, ground_truth = samples[index].image, samples[index].mask
```

Sample the ground truth mask sparsely:

```python
foreground_indices = np.nonzero(ground_truth.flatten())[0]
background_indices = np.nonzero(~(ground_truth.flatten()))[0]
```

```python
coverage = 0.01
```

```python
label = np.zeros(ground_truth.shape, dtype=np.int)
```

```python
random_foreground_indices = foreground_indices[np.random.randint(0, len(foreground_indices), coverage * len(foreground_indices))]
random_background_indices = background_indices[np.random.randint(0, len(background_indices), coverage * len(background_indices))]
```

```python
label.flat[random_foreground_indices] = 1
label.flat[random_background_indices] = -1
```

Form the label and strength arrays, which drive the segmentation.

```python
strength = np.zeros_like(img, dtype=np.float64)
strength[np.nonzero(label)] = 1.0
```

Run the growcut algorithm, the python implementation is *slow*.

```python
mask = growcut_cy.growcut(color.gray2rgb(img), np.dstack((label, strength)), window_size=4)
```

Plot the resulting segmentation.

```python
fig, ((ax1, ax2, ax3),(ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(12, 8))
ax1.imshow(img, interpolation='nearest', cmap='gray')
_ = ax1.axis('off')
ax1.set_title('natural image')

ax2.imshow(label, interpolation='nearest', cmap='gray')
_ = ax2.axis('off')
ax2.set_title('growcut labels')

ax3.imshow(strength, interpolation='nearest', cmap='gray')
_ = ax3.axis('off')
ax3.set_title('growcut label strength')

ax4.imshow(ground_truth, interpolation='nearest', cmap='gray')
ax4.set_title('human segmentation')
_ = ax4.axis('off')

ax5.imshow(mask == 1, interpolation='nearest', cmap='gray')
ax5.set_title('growcut segmentation')

_ = ax5.axis('off')
ax6.imshow((mask==1).astype(int) + ground_truth.astype(int), interpolation='nearest', cmap='jet')
_ = ax6.axis('off')
ax6.set_title('segmentation error')

plt.tight_layout()
```

```python
growcut_mask = mask == 1
```

```python
tp = float(sum( ground_truth == growcut_mask ))
fp = float(sum( ground_truth != growcut_mask ))
tn = float(sum( ~ground_truth == ~growcut_mask ))
fn = float(sum( ~ground_truth != ~growcut_mask ))
```

```python
precision = tp / (tp + fp)
recall = tp / (tp + fn)
accuracy = (tp + tn) / (tp + tn + fp + fn)
```

```python
F = 2 * ((precision * recall) / (precision + recall))
F
```
