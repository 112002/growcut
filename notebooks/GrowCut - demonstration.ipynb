GrowCut
-------

As described in:

    V. Vezhnevets, V. Konouchine. "Grow-Cut" - Interactive Multi-Label N-D Image Segmentation". 
    In Proceedings of the 2005 Conference, Graphicon. Pages 150 – 156.


```python
from growcut import growcut_cy
```

```python
import numpy as np
from matplotlib import pyplot as plt
from skimage import color
```

Form a simple example, which is a foreground "box" with some additive gaussian noise.

```python
grey = np.zeros((100, 100))
grey[:] = 0.5
grey[30:70, 30:70] = 0.51
grey += 0.02 * np.random.rand(*grey.shape)
grey = np.clip(grey, 0, 1)

image = color.gray2rgb(grey)
```

```python
fig, ax = plt.subplots(1,1, figsize=(5,4))
img = ax.imshow(image, interpolation='nearest')
plt.tight_layout()
```

Form the label and strength arrays, which drive the segmentation.

```python
label = np.zeros_like(grey, dtype=np.int)
label[:] = 0 # no label

label[0:10, 0:10] = -1 # background
label[90:99, 90:99] = -1 # background
label[90:99, 0:10] = -1 # background
label[0:10, 90:99] = -1 # background

label[45:55, 45:55] = 1 # foreground

strength = np.zeros_like(grey, dtype=np.float64)
strength[np.nonzero(label)] = 1.0
```

Visualize the labels and the image.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(image, interpolation='nearest', cmap='gray')
ax1.contour(label, colors='r')
ax1.axis('off')

img = ax2.imshow(label, interpolation='nearest', cmap='gray', vmin=-1, vmax=1)
_ = ax2.axis('off')
```

Run the growcut algorithm, the python implementation is *slow*.

```python
mask = growcut_cy.growcut(image, np.dstack((label, strength)), window_size=10)
```

Plot the resulting segmentation.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(image, interpolation='nearest', cmap='gray')
ax1.contour(mask, colors='r')
ax1.axis('off')

img = ax2.imshow(mask, interpolation='nearest', cmap='gray', vmin=-1, vmax=1)
_ = ax2.axis('off')
```
