Thresholding
------------

* Observe the distribution of the data and classify foreground and background based on a threshold.

```python
def threshold(level, component):
    return np.where(component < level, 1., 0.)
```

```python
mask = threshold(kangaroo[:,:,1], 0.7).astype(np.bool)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 4))
ax1.imshow(kangaroo); _ = ax1.axis('off')
ax2.imshow(~mask, cmap='binary'); _ = ax2.axis('off')
```

Local Thresholding
------------------

* Observe the local distribution of the data and classify foreground and background based on a threshold.

```python
def kernel(central_pixel, neighbours):   
   
    return np.allclose(neighbours[:,:,1], 0.8, atol=0.3) 
    
```

```python
def image_iterator(image):
    
    window_size = 2
    
    M, N = image.shape[0], image.shape[1]
    
    mask = np.zeros((M, N)).astype(np.bool)
    
    for m in range(M):
        for n in range(N):
                         
                neighbours = image[
                    max(0, m - window_size):min(m + window_size + 1, M),
                    max(0, n - window_size):min(n + window_size + 1, N),
                    :]
                
                central_pixel = image[m, n, :]
                
                mask[m, n] = kernel(central_pixel, neighbours)
    return mask
```

```python
local_mask = image_iterator(kangaroo)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 4))
ax1.imshow(kangaroo); _ = ax1.axis('off')
ax2.imshow(~local_mask, cmap='binary'); _ = ax2.axis('off')
```
