```python
from skimage import io, transform
```

```python
kangaroo = io.Image(transform.resize(io.imread('../examples/kangaroo.jpg'), (200, 400)))
```

Thresholding
============

* Observe the distribution of the data and classify foreground and background based on a threshold.

```python
def threshold(level, component):
    return np.where(component < level, 1., 0.)
```

```python
mask = threshold(kangaroo[:,:,1], 0.7).astype(np.bool)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 4))
ax1.imshow(kangaroo); _ = ax1.axis('off')
ax2.imshow(~mask, cmap='binary'); _ = ax2.axis('off')
```

```python
%timeit mask = threshold(kangaroo[:,:,1], 0.7).astype(np.bool)
```

Local Thresholding
==================

* Observe the local distribution of the data and classify foreground and background based on a threshold.

```python
def kernel(neighbours, threshold=0.8, tolerance=0.3):   
    return np.allclose(neighbours[:,:,1], threshold, atol=tolerance) 
```

```python
def image_iterator(image):
    
    window_size = 2
    
    M, N = image.shape[0], image.shape[1]
    
    mask = np.zeros((M, N)).astype(np.bool)
    
    for m in range(M):
        for n in range(N):
                         
                neighbours = image[
                    max(0, m - window_size):min(m + window_size + 1, M),
                    max(0, n - window_size):min(n + window_size + 1, N),
                    :]
                
                central_pixel = image[m, n, :]
                 
                mask[m, n] = kernel(neighbours)
    return mask
```

```python
local_mask = image_iterator(kangaroo)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 4))
ax1.imshow(kangaroo); _ = ax1.axis('off')
ax2.imshow(~local_mask, cmap='binary'); _ = ax2.axis('off')
```

```python
%timeit local_mask = image_iterator(kangaroo)
```

Optimization with Numba
=======================

```python
from numba import autojit
```

```python
@autojit
def window_floor(idx, radius):
    if radius > idx:
        return 0
    else:
        return idx - radius

@autojit
def window_ceil(idx, ceil, radius):
    if idx + radius > ceil:
        return ceil
    else:
        return idx + radius

@autojit
def kernel(image, i, j, window_radius, threshold, tolerance):
    
   height, width = image.shape[0], image.shape[1]
    
   for jj in xrange(window_floor(j, window_radius), window_ceil(j+1, width, window_radius)):
        for ii in xrange(window_floor(i, window_radius), window_ceil(i+1, height, window_radius)):
                      
            if (ii == i and jj == j):
                continue
            
            if abs(image[ii,jj,1] - threshold) > tolerance:
                return 0.0
    
   return 1.0

@autojit
def image_iter_kernel(image, window_radius, threshold=0.8, tolerance=0.3):
    
    height, width = image.shape[0], image.shape[1]
    
    mask = np.empty((height, width), dtype=np.bool)
    
    for j in xrange(width):
        for i in xrange(height):
                      
            mask[i,j] = kernel(image, i, j, window_radius, threshold, tolerance)
                        
    return mask
```

```python
mask_kernel_numba = image_iter_kernel(kangaroo, 2, 0.8, 0.3)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 4))
ax1.imshow(kangaroo); _ = ax1.axis('off')
ax2.imshow(~mask_kernel_numba, cmap='binary'); _ = ax2.axis('off')
```

```python
%timeit image_iter_kernel(kangaroo, 1, 0.8, 0.3)
```
